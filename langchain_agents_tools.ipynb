```python
import numpy as np
import pandas as pd
```

```python
from kaggle_secrets import UserSecretsClient
hfapi = UserSecretsClient().get_secret("HUGGING_FACE_API_KEY")
serpapi = UserSecretsClient().get_secret("SERP_API_KEY")
```

```python
import os
os.environ["HUGGINGFACEHUB_API_TOKEN"] = hfapi
os.environ["SERPAPI_API_KEY"] = serpapi
```

```python
from langchain.agents import load_tools
from langchain.agents import initialize_agent
from langchain.agents import AgentType
from langchain_community.llms import HuggingFaceEndpoint
from langchain.utilities import PythonREPL
from langchain_experimental.tools import PythonREPLTool

repo_id = "mistralai/Mistral-7B-Instruct-v0.2"

llm = HuggingFaceEndpoint(
    repo_id=repo_id, max_length=128, temperature=0.5, token=hfapi
)

python_repl = PythonREPL()
repl_tool = Tool(
    name="python_repl",
    description="A Python shell. Use this to execute python commands. Input should be a valid python command. If you want to see the output of a value, you should print it out with `print(...)`.",
    func=python_repl.run,
)

tools = load_tools(["wikipedia","serpapi","terminal"], llm=llm, allow_dangerous_tools=True)
tools.append(repl_tool)
```

```python
dogroom = initialize_agent(
    tools, llm, agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION, verbose=True, max_execution_time=500000, handle_parsing_errors=True
)
```

```python
dogroom.run("Create a whatsapp bot in python to delete whatsapp links in groups")
```
